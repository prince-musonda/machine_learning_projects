# Image caption generation Model

```python
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
```

```python
image_processor = BlipProcessor.from_pretrained("salesforce/blip-image-captioning-large")
model = BlipForConditionalGeneration.from_pretrained("salesforce/blip-image-captioning-large")
```

```python
def caption_image(image: Image) ->None:
  processed_image = image_processor(image, return_tensors='pt')
  processed_image = processed_image['pixel_values']
  tokenized_caption = model.generate(processed_image)
  text_caption = image_processor.decode(tokenized_caption[0],skip_special_tokens=True)
  return text_caption
```

```python
from torchvision import transforms
image = Image.open('/content/premium_photo-1710965560034-778eedc929ff.jpg')
image = transforms.Resize(256)(image)
image
```

```python
caption = caption_image(image)
print(caption)
```
